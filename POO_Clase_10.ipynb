```python
class Casa:
    """
    Contruir el constructor
    argumentos:
        color(str):color de la casa
        numero_habitaciones(int): el nmero de habitaciones de la casa
        numero_bao(int): el nmero de baos de la casa
        estacionamiento(bool): si tiene estacioaminto o no
        tiene_patio(bool): si tiene patio o no
        material(str): material de la casita
    """
    def __init__(self, color, numero_habitaciones, numero_bao, estacionamiento, tiene_patio,material):
        self.color = color
        self.numero_habitaciones = numero_habitaciones
        self.numero_bao = numero_bao
        self.estacionamiento = estacionamiento
        self.tiene_patio = tiene_patio
        self.material= material

    def descripcion(self):
        #Describur la casa
        descripcion=f"la casa es de color {self.color}, tiene { self.numero_habitaciones} habitacion(es), {self.numero_bao} bao(s), esta contruida en {self.material}"
        if self.estacionamiento:
            descripcion += " y tiene estacionamiento"

        if self.tiene_patio:
            descripcion += " y tiene patio"
        return descripcion

    def pintar(self,nuevo_color):
        self.color=nuevo_color
        print(f"la casa ha sido pintada de color {self.color}")
    def material(self,nuevo_material):
        self.material=nuevo_material
        print(f" la casa esta contruida en {self.material}")
```
