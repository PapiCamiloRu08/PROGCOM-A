Las matemáticas financieras son una rama aplicada de las matemáticas que facilita el análisis y la toma de decisiones en contextos económicos y empresariales. #Tasas de interes
Mide el valor de los intereses en un porcentaje para un periodo de tiempo (n) determinado. A la tasa de interes de denomina como "precio del dinero" ya que mide cuánto le cuesta a un agente usar el dinero.

 1. **Efectiva y/o periodica:** Indican verdaderamnete cuál es la rentabilidad de una inversion (Es el que se usa para las opeeraciones comerciales) o cuál es el crédito por periodo (n).
 2. **Nominal:** Son anuales con capitalización. No indican directamente la rentabilidad de una inversion (No se usa comercialmente).
 3. **Vencidas:** Cuando se liquidan al final del periodo (Todas las tasas son vencidas, al menos, que se exprese lo contrario (Anticipada)).
 4. **Anticipada:** El interes se liquida al comienzo del periodo.
 5. **Discreta:** Se aplican cuando el periodo de capitalización se mide en intervalos fijos de tiempo. (Todas las tasas son discretas al menos que se exprese lo contrario(Continua)).
 6. **Continua:** Se aplica en mercados bursatiles


 Con lo anteriror podemos concluir que, todas las tasas son periodicas, vencidas y discretas o efectivas anuales, almenos que se exprese lo contrario.

#Nomenclatura
1. **Efectiva Anual** = Siempre se vaa simbolizar como E.A o e.a es la unica que tiene execpcion en mayusculas

2. **Periodica** = Las periodicas siempre se simbolizaran como (mes.v) y su forma de leerse es mes = mensual


```
Semestre = Semestral = s.v
Trimestre = Trimestral = t.v
Bimestre = Bimestral = b.v
Meses = Mensual = m.v
Quincenal = Quincena = q.v
```
3. **Nominal:** Se indicaran en mayusculas y se le puede agregar la N al comienzo



```
- Semestre(s.v) = S.V = N.S.V = N.A.C.S.V
- Meses(m.v) = M.V = M.V = N.A.M.V
```
3. **Discreta:** Esta no tiene un simbolo como tal, y se sabe que todas las tasas son discretas almenos qu ese diga que es continua
4. **Continua:** Esta ira indicada en mayuscula tambien y debe terminar con la letra C al final


```
A.C = Anual Continua (Nominal)
N.A.C.T.A = Nominal Anual Capitalizable Trimestre Anticipado (No es continua)
```

````python
print(" Tengo el 34% C.C quiero saber cuál es su equivalencia en => E.A")
c= input("¿Cuál es su equivalencia?(Escribe cuatro decimales depues del punto y expresalo en porcentaje 1.2345%): " )
try:
    if c.endswith('%'):
        c_float = float(c[:-1])
    else:
        c_float = float(c)
    if c_float == 40.4947:
        print("Correcto")
    else:
        print("INCORRECTOO")
except ValueError:
    print("Invalid input. Please enter a number.")
````
